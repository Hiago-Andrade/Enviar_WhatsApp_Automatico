### Fazendo alterações nos dados

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']  # criando uma nova coluna chamda "Total"

df.head()
```

```python
# Mudando nome da Coluna "Name" e "Legendary"

df.rename(columns={"Name":"Nome","Legendary":"Lendário"}) #, inplace = True (para modificar no DataFrame original)
```

```python
# Excluir todas as linhas com valor Vazio

df.dropna()
```

```python
# Substituindo dados faltantes

df["Type 2"].fillna("Tipo desconhecido",inplace=True)                # Substitui por uma string
df["Type 2"].fillna(df["HP"].mean(),inplace=True)                   # Substitui pela média da coluna

df.head(2)
```

### Salvando nossos dados (exportando para o formato desejado)

```python
df.to_csv('modified.csv', index=False)

df.to_excel('modified.xlsx', index=False)

df.to_csv('modified.txt', index=False, sep='\t')
```

## Filtrando Dados

```python
df.loc[df['Type 1'] == "Grass"]  # filtrando na coluna "Type" apenas o tipo "Glass"

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')] # filtro "Type 1" é "Grass" ou "Type 2" é "Poison"

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # filtrando na coluna "Type" o tipo "Glass", "Type 2" apenas "Poison" e com "HP" maior que 70 

df.loc[df['Name'].str.contains('Mega')] # filtro os nomes que contem a palavra "Mega"

df.loc[~df['Name'].str.contains('Mega')] # filtro os nomes que não contem a palavra "Mega"

df.loc[df["Name"]!="Bulbasaur"] # Filtrando o Nome diferente de "Bulbasaur"
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df.reset_index(drop=True, inplace=True)  # baixando o novo banco de dados resetando o index(coluna a esquerda da #)

new_df.to_csv('filtro.csv')
```

## Estatísticas agregadas (Groupby)

```python
df.groupby(['Type 1']).mean() # organizando a média pelo "Type 1" (posso usar o count e o sum tambem)
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)#['Attack']  # Organizando da maior para menor média de ataque pelo "Type 1"
```

```python
df.groupby(['Type 1', 'Type 2']).mean()['Attack']  # Tabela dinamica com o grupo "Type 1" subgrupo "Type 2" pela média do ataque, mostrando só a coluna Ataque
```
