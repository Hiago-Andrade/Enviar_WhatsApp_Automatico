### 4 - Criando gráficos de linha

```python
import numpy as np

x1=np.arange(-1000,1000,1)
plt.plot(x1,-x1**3+4)
plt.show()
```

```python
### Gráfico mais avançado 

x = [0,1,2,3,4]
y = [0,2,4,6,8]

# # Redimensione seu gráfico (dpi especifica pixels por polegada, tamanho total. Figsize o tamanho das variaveis x e y)
plt.figure(figsize=(6,6), dpi=100)

# Linha 1

# Notação do argumento da linha "2x"
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Notação taquigrafia
# fmt = '[cor][marcador][linha]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# selecionando intervalo onde queremos plotar os pontos da linha vermelha
x2 = np.arange(0,4.5, 0.5)

# Plot parte do gráfico como linha
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# restante do gráfico como um ponto
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Adicionar um título (especificar parâmetros de fonte)
plt.title('Titulo de Exemplo', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 10})

# Subtitulo de X e de Y
plt.xlabel('Titulo variável X')
plt.ylabel('Titulo variáveil Y')

# Escala das variavel X e Y
plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10,12,14])

# Adicionar legenda
plt.legend()

# Salvar a figura (dpi 300 é bom ao salvar para que o gráfico tenha alta resolução)
plt.savefig('gráficoexemplo.png', dpi=300)

plt.show()
```

### 8 - Criando varios gráficos na mesma "Figure"

```python
y1 = x1**2
y2 = x1**3
y3 = -x1**2


plt.subplot(2,2,1)  # 2 linas e 2 colunas, colar na 1º parte
plt.plot(x1,y1)
plt.subplot(1,2,2)
plt.plot(x1,y2)
plt.subplot(2,2,3)
plt.plot(x1,y3)
```

```python
# Elevando o tamanho dos gráficos

y1 = x1**2
y2 = x1**3
y3 = -x1**2

tamanho = plt.figure(figsize=(8,8)) # Tamanho 8 da largura e 8 da altura

tamanho.add_subplot(2,2,1)  # 2 linas e 2 colunas, colar na 1º parte
plt.plot(x1,y1)
tamanho.add_subplot(1,2,2)
plt.plot(x1,y2)
tamanho.add_subplot(2,2,3)
plt.plot(x1,y3)
```

```python
# Fazendo um Lop para gerar os graficos

y4 = -x1**3

y = [y1,y2,y3,y4]

tamanho = plt.figure(figsize=(8,8))

for i in range(4):
    tamanho.add_subplot(2,2,i+1)
    plt.plot(x1,y[i])
```

```python
### Gráficos mais avançados 

h = np.arange(0,10,0.1)
fig1, f1_axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15,10))
fig1.suptitle("Demosntração Gráfica", size = 30)

f1_axes[0, 0].plot(np.sin(h), label="seno(h)", color = "purple")
f1_axes[0, 0].plot(np.cos(h), label="coss(h)")
f1_axes[0, 0].set_title("Relação Seno x Cosseno")
f1_axes[0, 0].set_xlabel("Subtitulo 1 - Label X")
f1_axes[0, 0].set_ylabel("Subtitulo 2 - Label y")

f1_axes[0, 1].plot(np.cos(h))
f1_axes[0, 1].set_title("Cosseno de h")

f1_axes[1, 0].plot(np.tan(h), color = "orange")
f1_axes[1, 0].set_title("Tangente de h")

f1_axes[1, 1].plot(np.sin(h), color = "purple")
f1_axes[1, 1].set_title("Seno de h")
```
